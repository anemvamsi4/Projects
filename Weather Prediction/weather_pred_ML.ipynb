#Balancing the Data & Labelling

```python
df['conditions'].value_counts()
```

```python
from sklearn.utils import resample

#Separate majority and minority classes
df_upsampled = df[df['conditions'] == 'Partially cloudy']
for con in ['Rain, Partially cloudy','Clear','Rain, Overcast','Overcast','Rain']:
  df_minority = df[df['conditions'] == con]

# Upsample minority class and other classes separately
# If not, random samples from combined classes will be duplicated and we run into
#same issue as before, undersampled remians undersampled.
  df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=6110,    # to match average class
                                 random_state=42) # reproducible results
 

# Combine majority class with upsampled minority class
  df_upsampled = pd.concat([df_upsampled, df_minority_upsampled])
```

```python
df_upsampled['conditions'].value_counts()
```

```python
X_ = df_upsampled.drop('conditions',axis = 1)
Y_con = df_upsampled['conditions']
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder 
le = LabelEncoder()
le.fit(Y_con)

Y_con = le.transform(Y_con)
```

```python
con = dict(zip(le.classes_, le.transform(le.classes_)))
con
```

```python
from sklearn.model_selection import train_test_split
xtrain_con,xtest_con,ytrain_con,ytest_con= train_test_split(X_,Y_con,random_state=0,test_size = 0.3)
```

#Model for condition: RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier(n_estimators = 55)
```

```python
res = model.fit(xtrain_con,ytrain_con)
```

```python
yp_con = model.predict(xtrain_con) 
accuracy1 = accuracy_score(yp_con,ytrain_con)
print('Train accuracy: ',accuracy1)
```

```python
yp_con = model.predict(xtest_con) 
accuracy2 = accuracy_score(yp_con,ytest_con)
print('Test accuracy: ',accuracy2)
```

```python
pickle.dump(model, open('con_model.pkl', 'wb'))
```

```python
from datetime import datetime
dateTime = datetime.now()
if dateTime.month<10:
  month = '0'+str(dateTime.month)
else:
  month = str(dateTime.month)
if dateTime.day<10:
  day = '0'+str(dateTime.day)
else:
  day = str(dateTime.day)
today = str(dateTime.year)+'-'+month+'-'+ day
today
```

```python
class Weather_pred:
  def __init__(self):
    self.data = None
    self.condition = None
    self.true_con = None
    self.true_des = None
    self.con_model = pickle.load(open('/content/drive/MyDrive/Projects/Weather Prediction/con_model.pkl', 'rb'))

  def get_data(self,Place,Date):
    req = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/" + Place + '/' + Date + '/' + Date + "?unitGroup=metric&include=days&key=J6J2CBFC25RTZ9CGFZTWYCQBD&contentType=csv"
    response = requests.request("GET", req)
    self.data = pd.DataFrame(csv.reader(response.text.splitlines(), delimiter=',',quotechar='"'))
  def show_data(self):
    print(self.data)
  
  def preprocess(self):
    input = self.data.drop(0)
    self.true_con = input[29][1]
    self.true_des = input[30][1]
    input = input.drop([31,32,26,27,28,30,29,0,1,13,14,22,23,24,25,15,16],axis=1)
    return input

  def predict(self,Place = 'nuzvid',Date = today):
    self.get_data(Place,Date)
    X = self.preprocess()
    con_out = self.con_model.predict(X)
    con = {'Clear': 0, 'Overcast': 1, 'Partially cloudy': 2, 'Rain': 3, 'Rain, Overcast': 4, 'Rain, Partially cloudy': 5}

    for key,values in con.items():
      if values== con_out:
        self.condition = key
        out1 = "Prediction: "+self.condition+"\nTruth: "+self.true_con
    out2 = self.true_des
    return out1,out2
```

```python
wp = Weather_pred()
```
